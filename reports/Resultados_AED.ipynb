# Análise Bivariada de Recursos Humanos

Este notebook apresenta uma análise bivariada de um dataset de recursos humanos, focando na relação entre várias características dos funcionários e a variável de saída `left` (saída da empresa).

**Objetivo**: Explorar as relações entre as variáveis do dataset e identificar padrões que possam explicar a saída dos funcionários.


```python
import pandas as pd
from IPython.display import Image, display
from pathlib import Path

# Carregar o sumário de distribuição
summary_path = Path("/content/Projeto_Demissao_Funcionario/reports/summary_distribution.csv")
summary_df = pd.read_csv(summary_path)

# Definir caminho para os gráficos
figures_path = Path("/content/Projeto_Demissao_Funcionario/reports/figures")

```

```python
# Exibir o gráfico de 'Satisfaction Level'
image_path = figures_path / "satisfaction_level_scatter_boxplot.png"
display(Image(filename=image_path))

```

### Funcionários que permaneceram (left = 0)

- **Mediana de satisfação:** Aproximadamente 0,7.
- **Intervalo interquartil (IQR):** De 0,57 a 0,82.
- **Distribuição:** Levemente assimétrica para baixo, com a maioria dos funcionários apresentando satisfação acima da mediana. Não há outliers, indicando que os dados estão dentro do esperado.

### Funcionários que saíram (left = 1)

- **Mediana de satisfação:** Aproximadamente 0,4.
- **Intervalo interquartil (IQR):** De 0,2 a 0,7.
- **Distribuição:** Cauda longa no lado inferior, sugerindo que alguns funcionários que saíram tinham níveis de satisfação muito baixos. Há maior variação e dispersão nos níveis de satisfação em comparação aos que permaneceram.

### Comparação geral

- Os funcionários que permaneceram têm níveis de satisfação significativamente mais altos do que os que saíram.
- A mediana de satisfação dos que permaneceram é quase o dobro da dos que saíram.
- Há uma zona de transição próxima a 0,6.


```python
# Exibir o gráfico de 'Work Accident'
image_path = figures_path / "Work_accident_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'Work_accident'
work_accident_summary = summary_df[summary_df['Variable'] == 'Work_accident']
print(work_accident_summary)
```

**Análise:** A priori, o trabalho em condições que levam a acidentes não se mostrou um fator tão impactante na saída dos funcionários. Entre aqueles que sofreram acidentes, a taxa de permanência é até mais alta (94,32%) do que entre aqueles que não sofreram acidentes. Isso pode indicar que a empresa oferece um bom suporte para os funcionários que se acidentam, o que contribui para uma maior retenção desses trabalhadores.

```python
# Exibir o gráfico de 'Average Monthly Hours'
image_path = figures_path / "average_montly_hours_scatter_boxplot.png"
display(Image(filename=image_path))
```

### Análise das Horas Médias Mensais Trabalhadas (average_monthly_hours)

#### Histograma
- **Funcionários que saíram da empresa (left = 1):** A distribuição é bimodal, com maior concentração em dois extremos: aqueles que trabalham entre 125-170 horas mensais e aqueles com mais de 240 horas mensais.
- **Funcionários que permanecem na empresa (left = 0):** Estão mais distribuídos, com picos entre 125-170 e 250-300 horas mensais.

#### Boxplot
- **Mediana das horas trabalhadas:** É mais alta para quem saiu da empresa (225 horas) em comparação aos que permaneceram (200 horas).
- **Variabilidade:** Há maior variabilidade nas horas trabalhadas entre aqueles que saíram, sugerindo que tanto trabalhar pouco quanto muito pode estar associado à saída.

#### Comparação Geral
Funcionários com cargas horárias moderadas (em torno de 200 horas mensais) são menos propensos a sair. Aqueles com cargas horárias extremas, seja muito baixas ou muito altas, têm maior probabilidade de deixar a empresa. Isso pode estar relacionado à insatisfação com a carga de trabalho, seja por subutilização ou excesso de trabalho.


```python
# Exibir o gráfico de 'Department'
image_path = figures_path / "department_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'department'
department_summary = summary_df[summary_df['Variable'] == 'department']
print(department_summary)
```

### Comparação Geral
- **Departamentos com Maior Retenção:** Management e RandD apresentam as maiores taxas de retenção, com mais de 87% dos funcionários permanecendo na empresa.
- **Departamentos com Maior Taxa de Saída:** RH, Accounting, e Technical mostram as maiores taxas de saída, com mais de 17% dos funcionários deixando a empresa.
- **Tendências:** Em geral, departamentos relacionados à gestão e pesquisa (Management, RandD) tendem a reter mais funcionários, enquanto departamentos administrativos como RH e Accounting têm uma maior taxa de saída. Isso pode indicar diferenças na satisfação ou nas condições de trabalho entre os departamentos.


```python
# Exibir o gráfico de 'Last Evaluation'
image_path = figures_path / "last_evaluation_scatter_boxplot.png"
display(Image(filename=image_path))
```

### Análise do Tempo desde a Última Avaliação (last_evaluation)

#### Histograma
- **Funcionários que saíram (left = 1):**
  - Maior concentração em dois grupos principais:
    - Aqueles com 0,47 a 0,6 anos desde a última avaliação.
    - Aqueles com 0,8 a 1,0 anos desde a última avaliação.
  - Funcionários com avaliações mais recentes (menos de 0,4 anos) têm uma maior probabilidade de permanecer na empresa.

#### Boxplot
- **Mediana do tempo desde a última avaliação:**
  - Maior para funcionários que saíram (~0,8 anos) em comparação aos que permaneceram (~0,7 anos).
- **Distribuição:**
  - O tempo desde a última avaliação para os que saíram é mais amplo, sugerindo que a falta de avaliações recentes pode estar associada à saída da empresa.

#### Considerações
Funcionários que não foram avaliados recentemente (entre 0,5 e 1 ano desde a última avaliação) têm maior probabilidade de sair, sugerindo que avaliações menos frequentes podem estar ligadas à insatisfação e à saída.



```python
# Exibir o gráfico de 'Number of Projects'
image_path = figures_path / "number_project_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'number_project'
number_project_summary = summary_df[summary_df['Variable'] == 'number_project']
print(number_project_summary)
```

### Análise do Número de Projetos (number_project)

#### Padrões Observados
- **Funcionários envolvidos em 3 a 4 projetos:**
  - **Permaneceram na empresa:** 98,92% (3 projetos) e 93,57% (4 projetos).
  - Esses funcionários são mais propensos a permanecer na empresa, sugerindo que esse nível de carga de trabalho é mais sustentável.

- **Funcionários envolvidos em 2 projetos:**
  - **Saíram da empresa:** 54,17%.
  - Apresentam uma maior taxa de saída, indicando que a subutilização pode levar à insatisfação e à rotatividade.

- **Funcionários envolvidos em 6 a 7 projetos:**
  - **Saíram da empresa:** 44,92% (6 projetos) e 100% (7 projetos).
  - Esses funcionários também estão mais propensos a deixar a empresa, o que sugere que uma sobrecarga de projetos pode ser insustentável e resultar em maior rotatividade.

#### Considerações
**Equilíbrio na Carga de Trabalho:** Funcionários com um número moderado de projetos (3 a 4) têm maior probabilidade de permanecer na empresa, com taxas de retenção de 98,92% e 93,57%, indicando que esse é um nível de carga de trabalho mais equilibrado e sustentável.

```python
# Exibir o gráfico de 'Promotion in Last 5 Years'
image_path = figures_path / "promotion_last_5years_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'promotion_last_5years'
promotion_summary = summary_df[summary_df['Variable'] == 'promotion_last_5years']
print(promotion_summary)
```

#### Considerações
A falta de promoção nos últimos 5 anos parece ser um fator significativo para a saída de funcionários, com 16,82% dos que não foram promovidos deixando a empresa. Em contraste, a promoção é um forte indicador de retenção, com 96,06% dos promovidos permanecendo na empresa. Embora a maioria dos funcionários que permaneceram também não tenha recebido promoção, a proporção é menor em comparação com aqueles que saíram, destacando a importância das oportunidades de crescimento para a retenção.

```python
# Exibir o gráfico de 'Salary'
image_path = figures_path / "salary_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'salary'
salary_summary = summary_df[summary_df['Variable'] == 'salary']
print(salary_summary)
```

### Análise dos Níveis Salariais (salary)

#### Funcionários com Salário Baixo
- **Saíram da empresa:** 20,45%
- **Permaneceram na empresa:** 79,55%
- O salário baixo está fortemente correlacionado com a decisão de sair, mas muitos ainda permanecem, indicando que outros fatores também influenciam a retenção.

#### Funcionários com Salário Médio
- **Saíram da empresa:** 14,62%
- **Permaneceram na empresa:** 85,38%
- A maioria dos funcionários com salários médios permaneceu, sugerindo que esse nível salarial é mais aceitável.

#### Funcionários com Salário Alto
- **Saíram da empresa:** 4,85%
- **Permaneceram na empresa:** 95,15%
- Poucos funcionários com salários altos saíram da empresa, mostrando uma forte correlação entre altos salários e maior retenção.


```python
# Exibir o gráfico de 'Time Spent in Company'
image_path = figures_path / "time_spend_company_countplot.png"
display(Image(filename=image_path))

# Exibir a distribuição percentual para 'time_spend_company'
time_spend_summary = summary_df[summary_df['Variable'] == 'time_spend_company']
print(time_spend_summary)
```

#### Considerações Gerais

### Principais Variáveis que Influenciam a Saída dos Funcionários

1. **Satisfaction Level (satisfaction_level)**
   - **Por quê:** Indicador chave de retenção e satisfação no trabalho, altamente relevante para prever a saída.

2. **Number of Projects (number_project)**
   - **Por quê:** Relacionado diretamente à carga de trabalho, com influência clara na decisão de permanência ou saída.

3. **Salary (salary)**
   - **Por quê:** Fator determinante na satisfação e retenção, essencial para entender a dinâmica de saída.

4. **Promotion Last 5 Years (promotion_last_5years)**
   - **Por quê:** A promoção está fortemente ligada à retenção, crucial para avaliar o avanço na carreira e satisfação.

#### Variáveis com Menor Potencial de Influência

1. **Work Accident (work_accident)**
   - **Por quê:** Mostrou menor impacto na decisão de saída, com alta retenção mesmo entre os acidentados.

2. **Department (department)**
   - **Por quê:** Diferenças entre departamentos são menos significativas, não sendo um fator central na decisão de saída.
